```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# a. Line Plot
plt.plot(x, y, color='blue', linestyle='-', marker='o')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()

# b. Area Plot
plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.title("Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# c. Scatter Plot
plt.scatter(x, y, color="red", marker='x')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# d. Bar Plot
categories = ['A','B','C','D']
values = [3,7,5,9]
plt.bar(categories, values, color="green")
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

# e. Histogram
data = np.random.randn(1000)   # 1000 random values
plt.hist(data, bins=30, color="purple", alpha=0.7)
plt.title("Histogram")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()

# f. Hexagonal Bin Plot
plt.hexbin(x, y, gridsize=30, cmap="Blues")
plt.title("Hexbin Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label="Counts")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
X = [1, 2, 4, 10, 11, 14]
Y = [5, 7, 9, 15, 18, 13]

# Line chart
plt.plot(X, Y, marker='o', linestyle='-', color='blue')

# Custom ticks on X-axis
plt.xticks([1, 4, 10, 14], labels=['One', 'Four', 'Ten', 'Fourteen'])

# Custom ticks on Y-axis
plt.yticks([5, 9, 15, 18], labels=['Low', 'Medium', 'High', 'Peak'])

# Titles and labels
plt.title("Line Chart with Custom Ticks")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Grid for clarity
plt.grid(True)

# Show plot
plt.show()
```
